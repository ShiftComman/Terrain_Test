```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python


# 
arr1 = np.random.rand(10000, 10000)
shape = arr1.shape

# 10000
idx = np.unravel_index(np.random.choice(np.prod(shape), 10000, replace=False), shape)

# 
plt.figure(figsize=(10, 10))
plt.scatter(idx[1], idx[0], alpha=0.5, marker='.')
plt.gca().invert_yaxis()  # (0,0)
plt.title("Distribution of Random Indices on the 2D Plane")
plt.xlabel("Second Dimension")
plt.ylabel("First Dimension")
plt.grid(True)
plt.show()
```

```python

# 
arr1 = np.random.rand(100, 100, 100)
shape = arr1.shape

# 10000
idx = np.unravel_index(np.random.choice(np.prod(shape), 10000, replace=False), shape)

# 
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

# 
axs[0].hist(idx[0], bins=range(shape[0]+1), edgecolor="k", align="left")
axs[0].set_title("Distribution along the 1st dimension")
axs[0].set_xlim(0, shape[0])
axs[0].set_xticks(range(shape[0]))

# 
axs[1].hist(idx[1], bins=range(shape[1]+1), edgecolor="k", align="left")
axs[1].set_title("Distribution along the 2nd dimension")
axs[1].set_xlim(0, shape[1])
axs[1].set_xticks(range(shape[1]))

# 
axs[2].hist(idx[2], bins=range(shape[2]+1), edgecolor="k", align="left")
axs[2].set_title("Distribution along the 3rd dimension")
axs[2].set_xlim(0, shape[2])
axs[2].set_xticks(range(shape[2]))

plt.tight_layout()
plt.show()
```
