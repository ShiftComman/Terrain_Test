```python

df = pd.DataFrame(pd.read_csv(r"D:\ArcgisData\pred_soildeep\deep_table\merges.csv"))

```

```python
df.columns
```

```python


# 使用groupby和agg计算每个组中E列的最大值
grouped = df.groupby(['横坐标_m_', '纵坐标_m_'])['取样截至深度_cm_'].agg('max').reset_index()

```

```python
grouped
```

```python
dfw = pd.merge(df,grouped,on=['横坐标_m_','纵坐标_m_'],how='left') 
```

```python
dfw
```

```python
dfw.columns
```

```python
duplicates = dfw.duplicated(subset=['横坐标_m_','纵坐标_m_','取样截至深度_cm__y'], keep='first')
df_duplicates = dfw[duplicates]
df_duplicates
```

```python
# 去除重复数据
dfw.drop_duplicates(subset=['横坐标_m_','纵坐标_m_','取样截至深度_cm__y'], keep='first', inplace=True)
```

```python
dfw.to_csv(r"D:\ArcgisData\pred_soildeep\deep_table\filter.csv")
```
