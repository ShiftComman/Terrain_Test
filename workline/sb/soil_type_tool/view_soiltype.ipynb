```python
# 查看土类分布
import arcpy
import pandas as pd
```

```python
# 训练点路径
train_point_path = r'F:\cache_data\shp_file\sb\sb_filter_points_train.shp'
```

```python
# df
df = pd.DataFrame(arcpy.da.FeatureClassToNumPyArray(train_point_path,['NEW_TL','NEW_YL','NEW_TS','NEW_TZ']))
```

```python
# 创建字典
soil_dict = {}

# 遍历DataFrame的每一行
for _, row in df.iterrows():
    # 将土类作为key，其他属性组成的列表作为value
    soil_dict[row['NEW_TZ']] = [row['NEW_TL'], row['NEW_YL'], row['NEW_TS'], row['NEW_TZ']]

# 如果需要去重
soil_dict = {k: v for k, v in soil_dict.items() if k is not None}

# 打印结果看看格式是否正确
print(soil_dict)
```

```python
len(soil_dict)
```

```python
# 按 NEW_TZ 分组并计算每个组的大小
result = df.groupby('NEW_TZ').size().reset_index(name='count')

# 筛选出计数少于 5 的 NEW_TZ
filtered_result = result[result['count'] < 5]

# 打印结果
print(filtered_result)
```

```python
# 获取四个标签的唯一值字典
unique_values_dict = {col: df[col].unique().tolist() for col in ['NEW_TL', 'NEW_YL', 'NEW_TS', 'NEW_TZ']}
# 获取每个标签的数量
count_dict = {len(unique_values_dict[col]): unique_values_dict[col] for col in unique_values_dict}
# 打印结果
[_ for _ in count_dict.items()]
```
