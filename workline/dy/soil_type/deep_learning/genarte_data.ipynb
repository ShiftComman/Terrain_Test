```python
import os
import pandas as pd
import numpy as np
```

```python
data_path = r'F:\cache_data\zone_ana\dy\train_data\train_20240619.csv'
```

```python
df = pd.read_csv(data_path)
```

```python
df.columns
```

```python
# 生成地块描述
def generate_description(df, columns_list):
    # 使用 pandas 的 apply 函数来创建描述列
    df['description'] = df.apply(lambda row: f"地块成土母质为{row[columns_list[1]]},土地利用类型为{row[columns_list[0]]}", axis=1)
    return df

```

```python
result_df = generate_description(df,['DLMC','母质'])
```

```python
features_list = ['description','TL','YL','TS','TZ',
 'Centroid_X',
 'Centroid_Y',
 'DEM_MEAN',
 'DEM_STD',
 'ndvi_MEAN',
 'PCA_0_MEAN',
 'Slope_MEAN',
 'Aspect_MEAN',
 'MRVBF_MEAN',
 'TopographicWetnessIndex_MEAN',
 'Mean_MEAN',
 'PH_MEAN',
 'PRE_MEAN',
 'SRA_MEAN',
 'TMP_MEAN',
 'VAP_MEAN',
 'WIND_MEAN',
 'PlanCurvature_MEAN']
```

```python
result_df = result_df[features_list]
```

```python
# 选择数值类型的列
import numpy as np
numeric_cols = result_df.select_dtypes(include=[np.float64])

# 计算数值列的均值
mean_values = numeric_cols.mean()

# 用均值填充数值列的空值
result_df[numeric_cols.columns] = numeric_cols.fillna(mean_values)
```

```python
result_df.to_csv(r'D:\model_path\DATA_PATH\result.csv',index=False)
```
